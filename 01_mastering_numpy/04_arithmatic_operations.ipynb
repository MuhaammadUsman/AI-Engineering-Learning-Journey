**<h1 align = "center"> Arithmetic Operations in NumPy </h1>**

# 1. Scalar Operations
- Perform arithmetic with a single number. 
- Every element receives the operation.

```python
# import numpy library
import numpy as np
```

```python
arr = np.array([2, 4, 6, 8])
```

```python
print("Addition:", arr + 2)  # Addition
print("Subtraction:", arr - 2)  # Subtraction
print("Multiplication:", arr * 2)  # Multiplication
print("Division:", arr / 2)  # Division  
print("Power:", arr ** 2) # Power
```

# 2. Vectorized Math Functions
- NumPy includes many mathematical functions.
- Example: sqrt, log, round, floor, ceil, exp, sin

```python
arr = np.array([2, 4, 6, 8, 10])

# 1. Square root
print("Square Root:", np.sqrt(arr))
# 2. rounding
arrs = np.array([1.2, 2.5, 3.7, 4.9])
print("Rounded:", np.round(arrs))
# 3. Floor 
print("Floor:", np.floor(arrs))
# 4. Ceil
print("Ceil:", np.ceil(arrs))
# 5. logarithm
print("Logarithm:", np.log(arr))
# 6. Exponential
print("Exponential:", np.exp(arr))
# 7. Power
print("Power:", np.power(arr, 2))
# 8. Absolute value
arrs = np.array([-1, -2, 3, -4])
print("Absolute Value:", np.abs(arrs))
# 9. Trigonometric functions
angles = np.array([0, 30, 45, 60, 90])
print("Sine:", np.sin(np.radians(angles)))
```

# 3. Element-wise Arithmetic

```python
arr1 = np.array([2, 4, 6])
arr2 = np.array([1, 3, 5])
# Element-wise addition
print("Element-wise Addition:", arr1 + arr2)
# Element-wise subtraction
print("Element-wise Subtraction:", arr1 - arr2)
# Element-wise multiplication
print("Element-wise Multiplication:", arr1 * arr2)
# Element-wise division
print("Element-wise Division:", arr1 / arr2)
```

# 4. Comparison Arithmetic

```python
marks = np.array([45, 55, 65, 75, 85])
print("Marks > 50:", marks > 50)  # Boolean array for marks greater than 50
```

```python
marks[0] = 95 # Modifying the first element
print("Modified Marks:", marks)
marks[marks < 80] = 0 # Setting marks less than 80 to 0
print("Marks after modification:", marks)
```

# 5. Real-World Reservoir Example
- This type of calculation is used daily in reservoir engineering.

```python
area = np.array([1000, 2000, 3000])
thickness = np.array([10, 12, 15])
porosity = np.array([0.15, 0.18, 0.12])

# Calculate pore volume
pore_volume = area * thickness * porosity
print("Pore Volume:", pore_volume)
```
