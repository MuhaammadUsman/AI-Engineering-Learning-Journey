**<h1 align = "center"> Descriptive Statistics </h1>**

### 1. Mean:
- The mean is the arithmetic average.
- Formula
    - Mean = ∑xi/n
- Example: GR = 45,48,50,47,49​​
- Mean: (45+48+50+47+49)/5 = 47.8

```python
import pandas as pd
# Creat dataframe 
df = pd.DataFrame({
    "Depth_m" : [1000, 1001, 1002, 1003, 1004, 1005],
    "GR_API": [45, 48, 50, 47, 49, 120],
    "Porosity_pct" : [18, 17, 16, 18, 17, 5],
    "Lithology" : ["Sandstone", "Shale", "Sandstone", "Limestone", "Siltstone", "Sandstone"]
})

print(df)
```

```python
# Mean of GR_API and Porosity_pct
print("GR Mean:", df["GR_API"].mean())
print("Porosity Mean:", df["Porosity_pct"].mean())
```

### 2. Median
- Median is the middle value after sorting.
- The value 120 pulls the mean upward.
- Median is less affected by outliers.

```python
# median calculation
print("Median of GR:", df["GR_API"].median())
print("Median of Porosity:", df["Porosity_pct"].median())
```

### 3. Mode
- The mode is the value that appears most frequently.
- Most common lithology, facies and mineral

```python
# Mode of lithology
print("Mode of Lithology:", df["Lithology"].mode())
```

### 4. Range:
- Range measures the spread between the maximum and minimum values.
- Formula
    - Range = Maximum − Minimum
- Small Range:
    - Consistant formation.
- Large Range:
    - Multiple lithologies or changing depositional environments.

```python
# range of GR
gr_range = df["GR_API"].max() - df["GR_API"].min()
print("Range of GR", gr_range)
```

### 5. Variance
- Variance measures how much the data varies around the mean.
- Formula:
    - s2 = ∑(xi​−xˉ)2/n−1
- High variance:
    - Highly heterogeneous reservoir
    - Variable lithology
    - Mixed sandstone and shale
- Low variance:
    - Uniform sandstone
    - Consistent carbonate interval

```python
# variance of GR_API
df["GR_API"].var()
```

### 6. Standard Deviation (STD)
- Standard deviation is the square root of the variance.
- STD = sqrt of Variance
- Low STD
    - Uniform lithology.

- High STD
    - Large lithological variation.

```python
# std of GR_API
print("STD:", df["GR_API"].std())
```

## Practical Comparison of Descriptive Statistics

| Statistic | What It Measures | Sensitive to Outliers? | Interpretation | Geological Applications |
|-----------|-------------------|------------------------|----------------|-------------------------|
| **Mean** | Average value of the dataset | ✅ Yes | Represents the overall average but can be influenced by extreme values. | Average Gamma Ray (GR), Porosity, Density (RHOB), Permeability |
| **Median** | Middle value after sorting | ❌ No | Represents the central value and is robust against outliers. | Typical well-log values, noisy logging data, skewed geological measurements |
| **Mode** | Most frequently occurring value | ❌ No | Identifies the most common observation or category. | Most common lithology, facies, mineral, or formation |
| **Range** | Difference between maximum and minimum values | ✅ Yes | Shows the total spread of the data. | Variation in Gamma Ray, Porosity, or Depth within an interval |
| **Variance** | Average squared deviation from the mean | ✅ Yes | Measures the overall variability or heterogeneity of the dataset. | Reservoir heterogeneity, lithological variability, petrophysical analysis |
| **Standard Deviation** | Average spread of values around the mean (same units as data) | ✅ Yes | Indicates how closely data points cluster around the mean. | Reservoir quality assessment, formation consistency, quality control (QC), anomaly detection |
