**<h1 align = "center"> DataFrame Operations in Pandas </h1>**

- 1. Merge DataFiles
- 2. Save DataFile
- 3. Create DataFrame
- 4. Export DataFrame (openpyxl install)

```python
# import pandas library
import pandas as pd
```

```python
# import data file
df = pd.read_csv("ecommerce_sales_data.csv")
df.head()
```

```python
# filter specific cols for file-1
df_cat = df[["Year", "Category", "Region"]]
# save in csv file
df_cat.to_csv("ProductsFile.csv", index = False)
# load file
df_prod = pd.read_csv("ProductsFile.csv")
df_prod.head()
```

```python
# filter specific cols for file-2
df_cat_sold = df[["Category", "Units_Sold", "Revenue", "Profit"]]
# save in csv file
df_cat_sold.to_csv("ProductsSold.csv", index = False)
# Load file
df_sold = pd.read_csv("ProductsSold.csv")
df_sold.head()
```

### 1. Merge Datafiles

```python
# merge two datafiles based on one similar col.
df_merged_files = pd.merge(df_prod, df_sold, on = "Category")
df_merged_files.head()
```

### 2. Save File

```python
# save merge file
df_merged_files.to_csv("Merged_File.csv", index = False)
```

### 3. Create DataFrames and Save in Excel Sheets
- This create with key value pair of dict.

```python
# create dataframe
df_cost = pd.DataFrame({
    "Category": ["Home", "Books", "Fashion"],
    "Quantit" : [200, 345, 455],
    "Cost": [3000, 4000, 1000]
    
})
print(df_cost)
```

```python
# 2nd DataFrame
df_revenue = pd.DataFrame({
    "Category" : ["Medical", "Electronics", "Home"],
    "Sold" : [123, 456, 500],
    "Revenue" : [1000, 35000, 4000]
})
print(df_revenue)
```

```python
# Save/Export in Excel Sheet
with pd.ExcelWriter("Cost_Revenue.xlsx") as writer:
    df_cost.to_excel(writer, sheet_name = "Cost", index = False)
    df_revenue.to_excel(writer, sheet_name = "Revenue", index = False)
```
