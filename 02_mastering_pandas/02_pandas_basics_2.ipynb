#### Dataset: Ecommerce Sales Dataset

```python
# import library
import pandas as pd
```

```python
# load the dataset
df = pd.read_csv("ecommerce_sales_data.csv")
```

```python
# display the first 5 rows of the dataset
df.head()
```

```python
# shape of the dataset (rows, codf.shape
df.shape
```

```python
# display column names
df.columns
```

```python
# unique values in the "Product Category" column
df["Category"].unique()
```

```python
# product category sales count
df["Category"].value_counts()
```

```python
# Filter required columns
df_filtered = df[["Year", "Category", "Units_Sold", "Revenue"]]
df_filtered.head()
```

```python
# filter the dataset for the year > 2021
df[df["Year"] > 2021]
```

```python
# filter dataset > 2021 and < 2024
df[(df["Year"] > 2021) & (df["Year"] < 2024)]
```

```python
# Filter category = "Books" and year = 2023
df[(df["Category"] == "Books") & (df["Year"] ==2023)]
```

```python
# describe the dataset
df.describe()
```

```python
# info about the dataset
df.info()
```

```python
# min and max ratings
print("Minimum Rating", df["Customer_Rating"].min())
print("Maximum Rating", df["Customer_Rating"].max())
```

```python
# create a new column "Revenue_per_Unit" by dividing "Revenue" by "Units_Sold" upto 2 decimal places
df["Revenue_per_Unit"] = (df["Revenue"] / df["Units_Sold"]).round(2)
```
