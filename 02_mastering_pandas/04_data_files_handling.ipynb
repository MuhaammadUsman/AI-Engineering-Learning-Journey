**<h1 align = "center"> Handling Data Files in Pandas </h1>**
- Rename cols
- Handle large dataset (rows)
- Replace with NaN values
- Create new cols
- Call custom code


```python
# import library
import pandas as pd
import numpy as np
# load dataset and rename cols with updates header and maxi 5 rows from dataset
df = pd.read_csv("ecommerce_sales_data.csv", nrows = 10,
                 names = ["Year", "Month", "Product_Name", "Region", "Units_Sold", "Revenue",
                          "Profit", "Rating", "Ad_Cost"], 
                 header = 1
                          )
```

```python
df
```

### Replace with NaN values

```python
# Replace with NaN values
df.replace(["N.A", "Not Available", -1.00, 999999.00], np.nan, inplace = True)
```

### Create New Col

```python
# create new cols
df["Tax"] = df["Revenue"] * 0.10
df.head()
```

### Save as csv 

```python
# save csv file 
df.to_csv("data_tax.csv", index = False)
```
